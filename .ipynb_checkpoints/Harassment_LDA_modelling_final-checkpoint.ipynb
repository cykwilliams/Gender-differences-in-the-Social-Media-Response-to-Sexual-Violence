```python
#13 Dec update: using spaCy rather than NLTK lemmatizer (as better) but this has to be applied to the full text (ie first)
#Also no longer need to expand contractions (though can do anyway) as spaCy does this for us
#N.B spaCy is a pipeline that does all of tokenizing, tagging, NER, lemmatizing etc for us (no stemmer as skips this step straight
#to lemmatize step which is more accurate, see https://towardsdatascience.com/text-normalization-with-spacy-and-nltk-1302ff430119)
#For some reason, the NLTK lemmatizer was generating the repeats in the tokenised list of text. Update: this is not true, it is the Phrases.Bigrams which is causing repeats!
# e.g [baffling, clapham, busy, even, night, cctv, everywhere, pounder, road, south, circular, someone, must, know, something, praying, found, soon, saraheverard, south_circular, south_circular, praying_found, south_circular, south_circular, praying_found, south_circular, south_circular, praying_found, south_circular, south_circular, praying_found, south_circular, south_circular, praying_found, south_circular, south_circular, praying_found, south_circular, south_circular, praying_found, south_circular]

####Topic modelling pre-processing
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet') 
import spacy
en_core = spacy.load('en_core_web_sm')


df_harass_selected_demog_VADER['full_text_all_spacylemmatized'] = df_harass_selected_demog_VADER['full_text_all']
df_harass_selected_demog_VADER["full_text_all_spacylemmatized"] = df_harass_selected_demog_VADER['full_text_all_spacylemmatized'].progress_apply(lambda x: " ".join([y.lemma_ for y in en_core(x)]))

#Replace new line
df_harass_selected_demog_VADER['full_text_all_processed']  = df_harass_selected_demog_VADER['full_text_all_spacylemmatized'] .str.replace("\n", " ",)
#Change "& amp" to "and"
df_harass_selected_demog_VADER['full_text_all_processed'] = df_harass_selected_demog_VADER['full_text_all_processed'].str.replace("& amp", "and")

#Removal hyperlinks
df_harass_selected_demog_VADER['full_text_all_processed']  = df_harass_selected_demog_VADER['full_text_all_processed'] .str.replace(r'https?://[^\s<>"]+|www\.[^\s<>"]+', " ")

#Remove emojis
#For full list of emojis to remove, see Denis da Mata entry here https://stackoverflow.com/questions/33404752/removing-emojis-from-a-string-in-python
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
def remove_emoji(string):
    return emoji_pattern.sub(r'', string)
df_harass_selected_demog_VADER['full_text_all_processed'] = df_harass_selected_demog_VADER['full_text_all_processed'].apply(remove_emoji) # Apply the remove_emoji function to each row in the text column

##Expand contractions
#List of contractions found in /home/chris/anaconda3/lib/python3.7/site-packages/contractions/data/contractions_dict.json
#Note that can't -> cannot, hence have to add space between
import contractions
df_harass_selected_demog_VADER['full_text_all_processed'] = df_harass_selected_demog_VADER['full_text_all_processed'].apply(contractions.fix)
#Now add space for cannot
df_harass_selected_demog_VADER['full_text_all_processed']  = df_harass_selected_demog_VADER['full_text_all_processed'] .str.replace("cannot", "can not")

#Remove punctuation (incl apostrophe) and non-standard characters, replace with space
df_harass_selected_demog_VADER['full_text_all_processed']  = df_harass_selected_demog_VADER['full_text_all_processed'] .str.replace('[^A-Za-z0-9 ]+', ' ')

#Removal of multiple (>2) occurrences of the same letter e.g lolll -> lol
df_harass_selected_demog_VADER['full_text_all_processed']  = df_harass_selected_demog_VADER['full_text_all_processed'] .str.replace("(.)\\1{2,}", "\\1",)
#Convert everything to lowercase
df_harass_selected_demog_VADER['full_text_all_processed']  = df_harass_selected_demog_VADER['full_text_all_processed'] .str.lower()

##Remove a list of stop words
#First need to tokenize:
from nltk.tokenize import word_tokenize
df_harass_selected_demog_VADER['full_text_all_processed_token']  = df_harass_selected_demog_VADER['full_text_all_processed'].progress_apply(word_tokenize)
#See list of stop words from  ntlk
stopwords.words('english')
#Now build custom list without belonging (I, you, we) words, negatives, verbs
stopwords_custom2 = ['pron', 'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're", "you've", "you'll", "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', "she's", 'her', 'hers', 'herself', 'it', "it's", 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', "that'll", 'these', 'those', 'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'don', "don't", 'should', "should've", 'now', 'd', 'll', 'm', 'o', 're', 've', 'y', 'ain', 'aren', "aren't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'ma', 'mightn', "mightn't", 'mustn', "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', "won't", 'wouldn', "wouldn't"]
df_harass_selected_demog_VADER['full_text_all_processed_no_stop2'] = df_harass_selected_demog_VADER['full_text_all_processed_token'].progress_apply(lambda x: [item for item in x if item not in stopwords_custom2])

#Lemmatize verbs into original stems
df_harass_selected_demog_VADER['full_text_all_processed_lemmatized'] = df_harass_selected_demog_VADER['full_text_all_processed_no_stop2'] 

#And save progress
df_harass_selected_demog_VADER.to_csv('~/Downloads/tweet-ids-001/SarahEverard/df_harass_selected_demog_VADER_orig_tweets_spacylemmatized.csv', sep=',', quotechar='"', quoting=csv.QUOTE_ALL, chunksize=100000)
```

##### Create LDA models, varying k number of topics, and visualise

```python
###Set k topic = 5

#create dictionary
text_dict_orig = Dictionary(df_harass_selected_demog_VADER_orig_tweets.full_text_all_processed_lemmatized)

#view integer mappings
text_dict_orig.token2id

#Create bag of words (bow)
tweets_bow_orig = [text_dict_orig.doc2bow(tweet) for tweet in df_harass_selected_demog_VADER_orig_tweets['full_text_all_processed_lemmatized']]
#List of 121194 generated, corresponding to the 121194 rows in df_harass_selected_demog_VADER_orig_tweets

#Fitting an LDA model in Gensim
from gensim.models.ldamodel import LdaModel


orig_tweets_lda5 = LdaModel(tweets_bow_orig,
                      num_topics = 5,
                      id2word = text_dict_orig,
                      random_state = 1,
                      chunksize=10000,
                      passes=10)

orig_tweets_lda5_vis = pyLDAvis.gensim_models.prepare(topic_model=orig_tweets_lda5, corpus=tweets_bow_orig, dictionary=text_dict_orig)
pyLDAvis.display(orig_tweets_lda5_vis)
```

## Classify individual Tweets by Topic from generated LDA model

```python
##Now load the topic models from file and classify first orig_tweets then _full:

def format_topics_sentences(ldamodel, corpus, texts, tweet_ids):
    # Init output
    sent_topics_df = pd.DataFrame()

    # Get main topic in each document
    for i, row in enumerate(ldamodel[corpus]):
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        # Get the Dominant topic, Perc Contribution and Keywords for each document
        print(i, 'i')
        for j, (topic_num, prop_topic) in enumerate(row):
            print(j, 'j')
            if j == 0:  # => dominant topic
                wp = ldamodel.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords]), ignore_index=True)
            else:
                break
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # Add original text to the end of the output
    contents = pd.Series(texts)
    tweet_unique_id = pd.Series(tweet_ids)
    print('About to concat')
    sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    sent_topics_df = pd.concat([sent_topics_df, tweet_unique_id], axis=1)
    print('format_topics_sentences finished')
    return(sent_topics_df)


def classify_topics(Mar_day_tweets_lda5, Mar_day_tweets_bow_orig, Mar_day):
    a = format_topics_sentences(Mar_day_tweets_lda5, Mar_day_tweets_bow_orig, Mar_day.full_text_all_processed_lemmatized.values.tolist(), Mar_day.id.values.tolist())
    # Format
    a.columns = ['Dominant_Topic', 'Topic_Perc_Contrib', 'Keywords', 'full_text_all_processed_lemmatized', 'id']
    print('About to merge')
    Mar_day_classified = Mar_day.merge(a, on='id')
    #Confirm same number of rows
    print(Mar_day.shape[0] == Mar_day_classified.shape[0], '; if True, then merge is successful')
    print((Mar_day_classified.full_text_all_processed_lemmatized_x != Mar_day_classified.full_text_all_processed_lemmatized_y).sum(), '; if 0, then full_text_all_processed_lemmatized column is the same in both dfs')
    print(Mar_day_classified.Dominant_Topic.isnull().sum(), 'rows without a dominant topic inferred')
    return Mar_day_classified
```

```python
##To classify using LDA model, need to first classify original df (i.e orig_tweets not NRC_summary), then merge based on 'full_text_all'
df_harass_selected_demog_VADER_orig_tweets_classified = classify_topics(orig_tweets_lda7g, tweets_bow_orig, df_harass_selected_demog_VADER_orig_tweets)
print(df_harass_selected_demog_VADER_orig_tweets_classified.shape)
#(121194, 59)
#Need to remove duplicate full_text_all_processed_lemmatized_x:
df_harass_selected_demog_VADER_orig_tweets_classified = df_harass_selected_demog_VADER_orig_tweets_classified.drop('full_text_all_processed_lemmatized_x',1)
df_harass_selected_demog_VADER_orig_tweets_classified = df_harass_selected_demog_VADER_orig_tweets_classified.rename(columns={'full_text_all_processed_lemmatized_y': 'full_text_all_processed_lemmatized'})

#Save progress
#df_harass_selected_demog_VADER_orig_tweets_classified.to_csv('~/Downloads/tweet-ids-001/SarahEverard/df_harass_selected_demog_VADER_orig_tweets_classified.csv', sep=',', quotechar='"', quoting=csv.QUOTE_ALL, chunksize=100000)
```
